1.The first task: breast-cancer-wisconsin


```
import numpy as np

#read the data
with open(TD_DIR+'breast-cancer-wisconsin.data', encoding='utf-8') as f:
  original_data = f.read().split('\n')
  for i in range(len(original_data)):
    original_data[i] = original_data[i].split(',')
#print(original_data[1:10,1:5])

#initialize the data
all_data_x = [[0 for col in range(9)] for row in range(len(original_data))]
all_data_y = [[0 for col in range(1)] for row in range(len(original_data))]

#in the data file 6th row has many '?'
for i in range(len(original_data)-1):
  if original_data[i][6] == '?':
    original_data[i][6] = 10

for i in range(len(original_data)-1):
  all_data_x[i][0] = int(original_data[i][1])
  all_data_x[i][1] = int(original_data[i][2])
  all_data_x[i][2] = int(original_data[i][3])
  all_data_x[i][3] = int(original_data[i][4])
  all_data_x[i][4] = int(original_data[i][5])
  all_data_x[i][5] = int(original_data[i][6])
  all_data_x[i][6] = int(original_data[i][7])
  all_data_x[i][7] = int(original_data[i][8])
  all_data_x[i][8] = int(original_data[i][9])
  all_data_y[i][0] = int(original_data[i][10])

#print(all_data_y)
 
```

Here is kNN part

```
from sklearn.model_selection import train_test_split 

train_X,test_X, train_y, test_y = train_test_split(all_data_x, all_data_y, test_size = 0.2, random_state = 24)

#result of knn 
result = [[0 for col in range(1)] for row in range(len(test_y))]

for i in range(len(test_X)):
  distance = [[0 for col in range(2)] for row in range(len(train_X))]

  #calcul the distance
  for j in range(len(train_X)):
    calcul = 0
    calcul = calcul+np.square(test_X[i][0]-train_X[j][0])+np.square(test_X[i][1]-train_X[j][1])+np.square(test_X[i][2]-train_X[j][2])
    calcul = calcul+np.square(test_X[i][3]-train_X[j][3])+np.square(test_X[i][4]-train_X[j][4])+np.square(test_X[i][5]-train_X[j][5])
    calcul = calcul+np.square(test_X[i][6]-train_X[j][6])+np.square(test_X[i][7]-train_X[j][7])+np.square(test_X[i][8]-train_X[j][8])
    calcul = np.sqrt(calcul)
    distance[j][0] = calcul
    distance[j][1] = train_y[j][0]

  #find the 5 minimal distances
  new_distance = sorted(distance, key=(lambda s:s[0]))

  if (new_distance[0][1]+new_distance[1][1]+new_distance[2][1]+new_distance[3][1]+new_distance[4][1])/5 <= 3:
    result[i] = 2
  else:
    result[i] = 4

#count the result
count = 0
for i in range(len(result)):
  if result[i] == test_y[i][0]:
    count = count + 1
print(count/len(result))
```

2.second task: haberman.data

```
import numpy as np

#read the data
with open(TD_DIR+'haberman.data', encoding='utf-8') as f:
  original_haberman_data = f.read().split('\n')
  for i in range(len(original_haberman_data)):
    original_haberman_data[i] = original_haberman_data[i].split(',')
#print(original_haberman_data)
#initialize the data
all_data_x = [[0 for col in range(len(original_haberman_data[0])-1)] for row in range(len(original_haberman_data))]
all_data_y = [[0 for col in range(1)] for row in range(len(original_haberman_data))]

for i in range(len(original_haberman_data)-1):
  for j in range(len(original_haberman_data[i])-1):
    all_data_x[i][j] = int(original_haberman_data[i][j])
  all_data_y[i][0] = int(original_haberman_data[i][len(original_haberman_data[i])-1])

from sklearn.model_selection import train_test_split 

train_X,test_X, train_y, test_y = train_test_split(all_data_x, all_data_y, test_size = 0.2, random_state = 25)

#result of knn 
result = [[0 for col in range(1)] for row in range(len(test_y))]

#set k = 5, meaning we take k plus near as reference
k = 5

for i in range(len(test_X)):
  distance = [[0 for col in range(2)] for row in range(len(train_X))]

  #calcul the distance
  for j in range(len(train_X)):
    calcul = 0
    for k in range(len(train_X[0])):
      calcul = calcul+np.square(test_X[i][k]-train_X[j][k])
    calcul = np.sqrt(calcul)
    distance[j][0] = calcul
    distance[j][1] = train_y[j][0]
  
  #find the 5 minimal distances
  new_distance = sorted(distance, key=(lambda s:s[0]))

  if sum(new_distance[0:k][1])/k <= 1.5:
    result[i] = 1
  else:
    result[i] = 2

#count the result
count = 0
for i in range(len(result)):
  if result[i] == test_y[i][0]:
    count = count + 1
print(count/len(result))
```
