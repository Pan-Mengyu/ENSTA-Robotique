```
import sys
#change the directory where you store the data
TD_DIR =  '/content/drive/My Drive/ROB311/TP4/'
sys.path.append(TD_DIR)
```

Read data and train

```
import pandas as pd
import numpy as np 

#read the data and make it a list
original_data = pd.read_csv(TD_DIR+"mnist_train.csv",header=None)
data = np.array(original_data.values.tolist())
print(type(data[0][0]))

#get train data
trainX = np.array([[data[row+1][col+1] for col in range(len(data[0])-1)] for row in range(len(data)-1)], dtype=int)
trainY = np.array([data[row+1][0] for row in range(len(data)-1)], dtype=int)

from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
pca = PCA(n_components=50, whiten=True)
svc = SVC(kernel='rbf', class_weight='balanced')
model = make_pipeline(pca, svc)

model.fit(trainX,trainY)
```

predict data

```
#read the data and make it a list
original_data_test = pd.read_csv(TD_DIR+"mnist_test.csv",header=None)
data_test = np.array(original_data_test.values.tolist())

testX = np.array([[data_test[row+1][col+1] for col in range(len(data_test[0])-1)] for row in range(len(data_test)-1)], dtype=int)
testY = np.array([data_test[row+1][0] for row in range(len(data_test)-1)], dtype=int)

predict_y = model.predict(testX)
count = 0

for i in range(len(testX)):
  if testY[i] == predict_y[i]:
    count = count + 1
print(count/len(testX))
```
